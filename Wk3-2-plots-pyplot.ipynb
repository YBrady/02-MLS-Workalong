# Two Variable Plots #
Adapted from https://matplotlib.org/tutorials/introductory/pyplot.html

```python
import matplotlib.pyplot as plt
```

```python
# 1-D plot
# Simple plot of y values against their position in array
plt.plot([1, 2, 3, 4]) 
```

```python
# 2-D Plot x vs y with line extrapolating in between values
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
```

```python
# 2-D plot without extrapolation - no line
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
# ro here means r for the colour (red)
# o is for the marker (circle) 
# Also can have .  : - (line) x  + etc
```

```python
import numpy as np
```

```python
# Creating an array using numpy
x = np.linspace(0.0, 100.0, 11) # Start, End , no of point
y1 = 2.0 * x + 5.0
y2 = 1.0 * x - 5.0
y3 = -2.0 * x + 0.0

plt.plot(x, y1,"r")
plt.plot(x, y2,"b")
plt.plot(x, y3,"g")
```
