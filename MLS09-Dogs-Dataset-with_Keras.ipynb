```python
# For neural networks.
import keras as kr
# For data frames.
import pandas as pd
# For numerical arrays.
import numpy as np
# For preprocessing data.
import sklearn.preprocessing as pre
# For splitting data sets.
import sklearn.model_selection as mod
# For whitening.
import sklearn.decomposition as dec
```

## Preparing

```python
df = pd.read_csv("data/dogs.csv")
df
```

Keras works best with numerical data rather than categorical. Best to encode categorical data to make it look like a number.

```python
# Week 9 method
# df["ispure"] = df["pure"].astype(np.int) # adding column where it is encoded

# Other way
# df["ispure"]=(df["pure"] == True).astype(np.int)
# df["isnotpure"] = (df["pure"] == False).astype(np.int)
# df

# Week 10 method - using the SKLearn Preprocessing binarizer
pureenc = pre.LabelBinarizer()
pureenc.fit(df['pure'])
df['pureenc'] = pureenc.transform(df['pure'])
```

```python
# Week 9
# Create the inputs for keras
# inputs = df[["age", "weight", "height", "ispure", "isnotpure"]]
# inputs

# Week 10
x = df[['pureenc', 'age', 'weight', 'height']]
y = df['lifespan']

x
```

```python
# Week 9
# And the outputs
# output = df["lifespan"]
# output

# Week 10
y
```

```python
# Split the dataset into training and testing
x_train, x_test, y_train, y_test = mod.train_test_split(x, y, test_size=0.3)
```

## Building and training

```python
m = kr.models.Sequential()

# Week 9
#m.add(kr.layers.Dense(10, input_dim=5, activation = "linear"))
#m.add(kr.layers.Dense(50, activation = "sigmoid"))
#m.add(kr.layers.Dense(10, activation = "linear"))
#m.add(kr.layers.Dense(10, activation = "linear"))
#m.add(kr.layers.Dense(1, activation = "linear"))

# Week 10
m.add(kr.layers.Dense(20, input_dim=4, activation = "relu"))
m.add(kr.layers.Dense(1, activation="linear"))

m.compile(loss = "mean_squared_error", optimizer ="adam")
```

```python
# Week 9
# m.fit(inputs, output, epochs=50, batch_size=10)

# Week 10
m.fit(x_train, y_train, epochs=100, batch_size=10)
```

```python
# Week 9
# m.predict(inputs)

# Week 10
m.predict(x_test).round().T
```

```python
# Week 9
# output.as_matrix()

# Week 10
y_test.as_matrix().astype(np.float32)
```

```python
# Week 9
# np.around(m.predict(inputs).T)

# Week 10
# Calculate the loss function (Mean squared error in our case)
m.evaluate(x_test, y_test)
```

## Scaling

```python
# Adding a scaler
scaler = pre.StandardScaler()
scaler.fit(x_train)
x_train_scaled = scaler.transform(x_train)
```

```python
# Rebuild the neural network
m = kr.models.Sequential()

m.add(kr.layers.Dense(20, input_dim=4, activation="relu"))
m.add(kr.layers.Dense(1, activation="linear"))

m.compile(loss="mean_squared_error", optimizer="adam")
```

```python
m.fit(x_train, y_train, epochs=100, batch_size=10)
```

```python
m.predict(x_test).round().T
```

```python
y_test.as_matrix().astype(np.float32)
```

```python
m.evaluate(x_test, y_test)
# Shows improvement from not standardised
```

## Whitening

```python
pca = dec.PCA(n_components=4, whiten=True)
pca.fit(x_train)
x_train_white = pd.DataFrame(pca.transform(x_train), columns=x.columns)
x_train_white
```

```python
m = kr.models.Sequential()

m.add(kr.layers.Dense(100, input_dim=4, activation="tanh"))
m.add(kr.layers.Dense(1, activation="linear"))

m.compile(loss="mean_squared_error", optimizer="adam")
```

```python
m.fit(x_train_white, y_train, epochs=100, batch_size=10)
```

```python
m.predict(pca.transform(x_test)).round().T
```

```python
y_test.as_matrix().astype(np.float32)
```

```python
m.evaluate(pca.transform(x_test), y_test)
```
