```python
# Import the packages
import matplotlib.pyplot as plt
import numpy as np
import keras as kr
```

```python
# Setting a plot size
plt.rcParams['figure.figsize'] = (10, 10)
```

```python
# Array of evenly spaced numbers
x = np.linspace(-10.0, 10.0, 2000)

# Creating 3 arrays
y_l = (3.0 * x) + 5.0 # y_l - linear 
y_p = (x * x) - 41.0 # y_p - polynomial
y_s = 10.0 * np.sin(2.0 * x) # y_s - sinoidal
```

```python
# PLot the 3 arrays
# See using Latex in labels!
plt.plot(x, y_l, label='$3x + 5$')
plt.plot(x, y_p, label='$x^2 - 41$')
plt.plot(x, y_s, label='$10 \sin(2x)$')

plt.legend()
```

```python
#y = y_l # linear - just use 1 neural netwurk 
y = y_p # 
#y = y_s # very difficult to train

# Create a new neural network.
m = kr.models.Sequential()

# Add neurons.
#m.add(kr.layers.Dense(1, input_dim=1, activation="linear"))

# Add neurons
# TO optimise you can change the 10 value and / or the activation
m.add(kr.layers.Dense(20, input_dim=1, activation="tanh")) #tanh output for p_l
## tanh always give values from 0 and 1 for positive y and 0 and -1 for negative 9
m.add(kr.layers.Dense(20, activation="sigmoid")) 
m.add(kr.layers.Dense(20, activation="tanh")) 
# Linear means no activation
m.add(kr.layers.Dense(1, activation='linear'))

# Compile the model.
# To optimise you could change loss rarely do
# Sometimes use entropyfrom mean squared error but best no to.
# Could change optimiser too 
m.compile(loss="mean_squared_error", optimizer="adam") # another optimiser - "sgd"
# sgd - stochastic gradient descent

# Train the model.
# To optimise you can always change the epochs and batch size
m.fit(x, y, epochs=50, batch_size=10)
# epochs = no of time to perform training
# batch_size = no of values to use each time

# Learning rate - number associated with size of jump a neural network 
# is willing to take when it gets something wrong.
# The neural network will see how far it is via mean_squared_error.
# Learning rate is usually the problem noted in optimizer. Can over-jump.
# In this instance adam is better than sgd

# Run each x value through the neural network.
p = m.predict(x)

# Plot the values.
plt.plot(x, p, 'r-', label='Prediction')
plt.plot(x, y, 'k:', label='Original')
plt.legend()
```
